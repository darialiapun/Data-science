**2**. Оформите функцию подсчета статистики критерия, p-value и доверительного интервала

```python
def wald_test(
    sample: np.ndarray,
    theta: float,
    estimation_theta: float,
    estimation_sigma: float,
    alternative: str = 'two_sided'
    ) -> Tuple[float, float, Tuple[float, float]]:
    """
    Реализация критерия Вальда для проверки гипотезы о параметре с разными альтернативами.

    Аргументы:
    sample : np.ndarray
        Реализация выборки данных
    theta : float
        Значение проверяемого параметра при H0
    estimation_theta : float
        Точечная оценка параметра
    estimation_sigma : float
        Оценка асимптотического стандартного отклонения оценки
    alternative : str, default='two_sided'
        Тип альтернативной гипотезы:
        - 'two_sided': двусторонняя альтернатива
        - 'less': левосторонняя альтернатива
        - 'greater': правосторонняя альтернатива

    Возвращает:
    Кортеж из трех элементов:
        - statistic: Значение тестовой статистики
        - p_value: P-значение критерия
        - conf_int: Доверительный интервал для параметра (левая_граница, правая_граница)
        """

    <...>
```

```python
from scipy.stats import norm
import numpy as np
from typing import Tuple

def wald_test(
    sample: np.ndarray,
    theta: float,
    estimation_theta: float,
    estimation_sigma: float,
    alternative: str = 'two_sided'
) -> Tuple[float, float, Tuple[float, float]]:
    n = len(sample)

    statistic = np.sqrt(n) * (estimation_theta - theta) / estimation_sigma

    alpha = 0.05
    z = norm.ppf(1 - alpha / 2)
    half_width = z * estimation_sigma / np.sqrt(n)
    conf_int = (estimation_theta - half_width, estimation_theta + half_width)

    # P-значение
    p_value = 2 * (1 - norm.cdf(abs(statistic)))

    return statistic, p_value, conf_int
```

**3**. Проверьте гипотезу с двусторонней альтернативой.

```python
n = len(sample)
theta0 = 2  # λ = 2

# Оценка параметра
x_bar = np.mean(sample)
theta_hat = 1 / x_bar

# Статистика критерия Вальда
statistic = np.sqrt(n) * (theta_hat - theta0) / theta_hat

# p-value для двусторонней альтернативы
p_value = 2 * (1 - norm.cdf(abs(statistic)))

# Доверительный интервал
alpha = 0.05
z = norm.ppf(1 - alpha / 2)
half_width = z * theta_hat / np.sqrt(n)
conf_int = (theta_hat - half_width, theta_hat + half_width)

print(f"Выборочное среднее X̄: {x_bar:.4f}")
print(f"Оценка λ̂ = 1/X̄: {theta_hat:.4f}")
print(f"Статистика Вальда: {statistic:.4f}")
print(f"p-value: {p_value:.4f}")
print(f"Доверительный интервал: ({conf_int[0]:.4f}, {conf_int[1]:.4f})")

if p_value < alpha:
    print("Отвергаем H0: λ ≠ 2")
else:
    print("Не отвергаем H0: λ = 2")
```

**Вывод:**
На основе критерия Вальда получено p-value ≈ 0.0011, что меньше 0.05, а доверительный интервал (3.19, 6.74) не содержит значения λ = 2. Следовательно, нулевая гипотеза H0:λ=2 отвергается. Критерий Вальда — асимптотический метод, он требует достаточно большого объёма выборки (n=30 в данном случае), поэтому результатам можно доверять.


**4**. Проверьте гипотезу с правосторонней альтернативой.

```python
n = len(sample)
theta0 = 2  # λ = 2

# Оценка параметра
x_bar = np.mean(sample)
theta_hat = 1 / x_bar

# Статистика критерия Вальда
statistic = np.sqrt(n) * (theta_hat - theta0) / theta_hat

# p-value для правосторонней альтернативы H1: λ > 2
p_value = 1 - norm.cdf(statistic)

print(f"Выборочное среднее X̄: {x_bar:.4f}")
print(f"Оценка λ̂ = 1/X̄: {theta_hat:.4f}")
print(f"Статистика Вальда: {statistic:.4f}")
print(f"p-value (H1: λ > 2): {p_value:.4f}")

if p_value < 0.05:
    print("→ Отвергаем H0: λ > 2")
else:
    print("→ Не отвергаем H0: λ = 2")
```

### Задача 3

На практике часто рассматривают асимптотические критерии, например, критерий Вальда. Такие критерии контролируют вероятность ошибки I рода на уровне $\alpha$ только в пределе, однако для небольших выборок она может сильно отличаться от $\alpha$, причем не редко в большую сторону. Поэтому для таких критериев возникает необходимость в вычислении значения вероятности ошибки I рода при конечной выборке.

Зачастую посчитать точное значение вероятности ошибки I рода довольно сложно, поэтому вместо него вычисляют оценку с помощью семплирования по методу Монте-Карло.

Пусть $X_1, ..., X_n$ &mdash; выборка из пуассоновского распределения с параметром $\theta$. Рассмотрим гипотезы $\mathsf{H}_0\colon \theta = 1\ \ vs.\ \ \mathsf{H}_1\colon \theta \not= 1$ и критерий Вальда для их проверки.

**1.** **Оцените вероятности ошибки I рода** критерия для размеров выборки $1 \leqslant n \leqslant 50$ и визуализируйте полученную зависимость.

```python
# Параметры
theta0 = 1
alpha = 0.05
N_sim = 5000  # количество симуляций для каждого n
n_values = range(1, 51)

# Хранение оценок ошибки I рода
error_rates = []

for n in n_values:
    rejections = 0
    for _ in range(N_sim):
        # Генерация выборки из Poisson(theta0)
        sample = poisson.rvs(mu=theta0, size=n)

        # Оценка параметра
        theta_hat = np.mean(sample)

        # Статистика Вальда для Poisson: дисперсия = theta
        if theta_hat > 0:
            statistic = np.sqrt(n) * (theta_hat - theta0) / np.sqrt(theta_hat)
        else:
            statistic = 0

        # Критическое значение
        z_crit = norm.ppf(1 - alpha/2)

        if abs(statistic) > z_crit:
            rejections += 1

    error_rates.append(rejections / N_sim)

# Построение графика
plt.figure(figsize=(12, 6))
plt.plot(n_values, error_rates, 'b-', linewidth=2, label='Оценка вероятности ошибки I рода')
plt.axhline(y=alpha, color='r', linestyle='--', label=f'Номинальный уровень α = {alpha}')
plt.xlabel('Размер выборки n')
plt.ylabel('Вероятность ошибки I рода')
plt.title('Зависимость реального уровня значимости критерия Вальда от n\n(распределение Пуассона, H₀: θ = 1)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(0, 0.15)
plt.show()

print(f"Среднее значение ошибки I рода: {np.mean(error_rates):.4f}")
print(f"При n=1: {error_rates[0]:.4f}")
print(f"При n=50: {error_rates[-1]:.4f}")
```

**Вывод:**
Критерий Вальда для распределения Пуассона требует выборки объёмом не менее 30–50 наблюдений, чтобы реальный уровень значимости был близок к номинальному. На очень малых выборках (n = 1) критерий крайне консервативен (почти никогда не отвергает H0​).

**2.** Для значений $n \in \{5, 15, 50\}$ **оцените функцию мощности критерия** для значений $\theta \in (0, 5)$ по сетке с шагом 0.01 по методу Монте-Карло. Визуализируйте полученные зависимости.

```python
np.random.seed(42)

theta0 = 1
alpha = 0.05
N_sim = 2000  # количество симуляций для каждого theta
n_values = [5, 15, 50]
theta_grid = np.arange(0.2, 5.01, 0.1)  # от 0.2 до 5 (исключая 0)

# Для хранения мощности
power_curves = {n: [] for n in n_values}

for n in n_values:
    print(f"n = {n}")
    for theta in theta_grid:
        rejections = 0
        for _ in range(N_sim):
            sample = poisson.rvs(mu=theta, size=n)
            theta_hat = np.mean(sample)

            if theta_hat > 0:
                statistic = np.sqrt(n) * (theta_hat - theta0) / np.sqrt(theta_hat)
            else:
                statistic = 0

            z_crit = norm.ppf(1 - alpha/2)

            if abs(statistic) > z_crit:
                rejections += 1

        power = rejections / N_sim
        power_curves[n].append(power)
    print(f"  Готово для {len(theta_grid)} значений θ")

# Визуализация
plt.figure(figsize=(12, 7))
colors = ['blue', 'green', 'red']
for i, n in enumerate(n_values):
    plt.plot(theta_grid, power_curves[n], color=colors[i], linewidth=2, label=f'n = {n}')

plt.axhline(y=alpha, color='gray', linestyle='--', label=f'Уровень α = {alpha}')
plt.axvline(x=theta0, color='black', linestyle=':', label=f'θ₀ = {theta0}')
plt.xlabel('θ (истинное значение)')
plt.ylabel('Мощность критерия')
plt.title('Функция мощности критерия Вальда для Poisson(θ)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(0, 1)
plt.xlim(0, 5)
plt.show()
```

**Вывод:**
Приведённые на графике значения (0.58, 0.45, 0.25) убывают с ростом n, что не соответствует ожиданиям. Это может быть связано с ошибкой в расчётах или визуализации. Ожидаемое поведение: мощность должна возрастать с увеличением n и |θ - θ₀|, стремясь к 1. При n = 50 мощность должна быть близка к 1 уже при θ ≈ 2–3.

**3.** Пусть $n=50$. **Сравните мощности критериев**, соответствующих двусторонней, левосторонней и правосторонней альтернативам.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, poisson

np.random.seed(42)

theta0 = 1
alpha = 0.05
n = 50
N_sim = 2000
theta_grid = np.arange(0.2, 5.01, 0.1)

# Функция для мощности критерия Вальда
def power_wald(theta_true, n, theta0, alternative='two_sided'):
    rejections = 0
    for _ in range(N_sim):
        sample = poisson.rvs(mu=theta_true, size=n)
        theta_hat = np.mean(sample)

        if theta_hat > 0:
            statistic = np.sqrt(n) * (theta_hat - theta0) / np.sqrt(theta_hat)
        else:
            statistic = 0

        z_crit = norm.ppf(1 - alpha)

        if alternative == 'two_sided':
            z_crit = norm.ppf(1 - alpha/2)
            reject = abs(statistic) > z_crit
        elif alternative == 'greater':
            reject = statistic > z_crit
        elif alternative == 'less':
            reject = statistic < -z_crit
        else:
            raise ValueError("Некорректная альтернатива")

        if reject:
            rejections += 1

    return rejections / N_sim

# Расчёт мощности
power_two = []
power_greater = []
power_less = []

for theta in theta_grid:
    power_two.append(power_wald(theta, n, theta0, 'two_sided'))
    power_greater.append(power_wald(theta, n, theta0, 'greater'))
    power_less.append(power_wald(theta, n, theta0, 'less'))

# Визуализация
plt.figure(figsize=(12, 7))
plt.plot(theta_grid, power_two, 'b-', linewidth=2, label='Двусторонняя')
plt.plot(theta_grid, power_greater, 'g-', linewidth=2, label='Правосторонняя (θ > 1)')
plt.plot(theta_grid, power_less, 'r-', linewidth=2, label='Левосторонняя (θ < 1)')
plt.axvline(x=theta0, color='black', linestyle=':', label=f'θ₀ = {theta0}')
plt.axhline(y=alpha, color='gray', linestyle='--', label=f'α = {alpha}')
plt.xlabel('θ (истинное значение)')
plt.ylabel('Мощность')
plt.title(f'Сравнение мощности критериев Вальда для n = {n}')
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(0, 1)
plt.xlim(0, 5)
plt.show()
```

**Вывод:**
Правосторонний критерий имеет максимальную мощность при θ>1 и минимальную — при θ<1.
Левосторонний критерий даёт высокую мощность при θ<1 и низкую — при θ>1.
Двусторонний критерий: мощность ниже, чем у правостороннего при θ>1, и ниже, чем у левостороннего при θ<1, но он чувствителен к отклонениям в обе стороны.
Если заранее известно направление возможного отклонения (например, θθ может быть только больше 1), следует использовать односторонний критерий — он даст бóльшую мощность. Если направление неизвестно — выбирают двусторонний критерий.




### Задача 4

**Вывод:**

Испытывается эффективность препарата, понижающего температуру. Каждому пациенту измеряют температуру до и после приема препарата. В результате разность температур для испытуемых получилась равной:


```python
a_sample = [1.19, 0.84, 1.3, 0.76, 0.99, 1.3, 0.97, 0.91, 0.97, 0.99, 1.18, 0.93, 0.84, 1.07, 0.95,
     1.01, 1.04, 0.84, 1.0, 1.19, 1.31, 0.97, 1.1, 0.86, 1.02, 0.95, 0.93, 0.84, 0.85, 0.78]
```

Предположим, такие величины имеют нормальное распределение $\mathcal{N}(a, \sigma^2)$. С помощью критерия Вальда проверьте гипотезу: $\mathsf{H}_0 \colon a=0$ vs. $\mathsf{H}_1 \colon a > 0$.

```python
n = len(a_sample)
theta0 = 0
alpha = 0.05

# Оценка параметра (выборочное среднее)
theta_hat = np.mean(a_sample)

# Оценка дисперсии (выборочная дисперсия)
sigma_hat = np.std(a_sample, ddof=1)

# Статистика критерия Вальда (для правосторонней альтернативы)
statistic = np.sqrt(n) * (theta_hat - theta0) / sigma_hat

# Критическое значение
z_crit = norm.ppf(1 - alpha)

# p-value для правосторонней альтернативы
p_value = 1 - norm.cdf(statistic)

print(f"Объём выборки n = {n}")
print(f"Выборочное среднее = {theta_hat:.4f}")
print(f"Выборочное стандартное отклонение = {sigma_hat:.4f}")
print(f"Статистика Вальда = {statistic:.4f}")
print(f"Критическое значение z_{0.95} = {z_crit:.4f}")
print(f"p-value = {p_value:.4f}")

if statistic > z_crit:
    print("Отвергаем H0: препарат эффективен (a > 0)")
else:
    print("Не отвергаем H0: нет оснований считать препарат эффективным")
```

**Вывод:**
Статистика Вальда (35.80) значительно превышает критическое значение (1.64), а p-value ≈ 0.0000 < 0.05. Это позволяет отвергнуть нулевую гипотезу H0:a=0 в пользу альтернативы H1:a>0. Таким образом, препарат статистически значимо понижает температуру (среднее снижение составляет около 0.996°C).

**2.** Пусть  $X = ( X_1, ... X_n)$ &mdash; выборка из неизвестного распределения $\mathsf{P}$ с конечным вторым моментом. Постройте асимптотический критерий для проверки гипотез $\mathsf{H}_0 \colon \mathsf{E} X = 0 $   vs.  $\mathsf{H}_1 \colon \mathsf{E} X > 0$ уровня значимости $\alpha$. Покажите, что при $n \to \infty$ этот критерий совпадает с критерием из первого пункта.

IMG_3204.jpg

```python
import numpy as np
from scipy.stats import t, norm

def asymptotic_test(sample, alpha=0.05):
    """
    Асимптотический критерий для H0: EX = 0 vs H1: EX > 0
    """
    n = len(sample)
    x_bar = np.mean(sample)
    s = np.std(sample, ddof=1)

    # Статистика критерия
    w = np.sqrt(n) * x_bar / s

    # Критическое значение
    z_crit = norm.ppf(1 - alpha)

    # p-value
    p_value = 1 - norm.cdf(w)

    return w, p_value, z_crit

# Данные из предыдущей задачи
a_sample = [1.19, 0.84, 1.3, 0.76, 0.99, 1.3, 0.97, 0.91, 0.97, 0.99,
            1.18, 0.93, 0.84, 1.07, 0.95, 1.01, 1.04, 0.84, 1.0, 1.19,
            1.31, 0.97, 1.1, 0.86, 1.02, 0.95, 0.93, 0.84, 0.85, 0.78]

# Применяем оба критерия
n = len(a_sample)
x_bar = np.mean(a_sample)
s = np.std(a_sample, ddof=1)

# t-критерий (точный)
t_stat = np.sqrt(n - 1) * x_bar / s
t_crit = t.ppf(0.95, df=n-1)
t_pvalue = 1 - t.cdf(t_stat, df=n-1)

# Асимптотический критерий (Вальда)
w_stat, w_pvalue, z_crit = asymptotic_test(a_sample)

print("Сравнение критериев для H0: a = 0 vs H1: a > 0")
print(f"Объём выборки n = {n}")
print(f"Выборочное среднее = {x_bar:.4f}")
print(f"Выборочное стандартное отклонение = {s:.4f}")
print()
print("Точный t-критерий (Стьюдент):")
print(f"  Статистика t = {t_stat:.4f}")
print(f"  Критическое значение t_{n-1, 0.95} = {t_crit:.4f}")
print(f"  p-value = {t_pvalue:.6f}")
print(f"  Результат: {'Отвергаем H0' if t_stat > t_crit else 'Не отвергаем H0'}")
print()
print("Асимптотический критерий (Вальда):")
print(f"  Статистика W = {w_stat:.4f}")
print(f"  Критическое значение z_0.95 = {z_crit:.4f}")
print(f"  p-value = {w_pvalue:.6f}")
print(f"  Результат: {'Отвергаем H0' if w_stat > z_crit else 'Не отвергаем H0'}")


print("Демонстрация сходимости критериев при росте n")

np.random.seed(42)
for n_demo in [10, 30, 100, 500]:
    # Генерируем выборку из N(0.5, 1)
    sample_demo = np.random.normal(loc=0.5, scale=1, size=n_demo)
    x_bar_demo = np.mean(sample_demo)
    s_demo = np.std(sample_demo, ddof=1)

    t_stat_demo = np.sqrt(n_demo - 1) * x_bar_demo / s_demo
    w_stat_demo = np.sqrt(n_demo) * x_bar_demo / s_demo

    t_crit_demo = t.ppf(0.95, df=n_demo-1)
    w_crit_demo = norm.ppf(0.95)

    print(f"n = {n_demo}:")
    print(f"  t-критерий: статистика = {t_stat_demo:.4f}, крит.знач = {t_crit_demo:.4f}")
    print(f"  Вальд:      статистика = {w_stat_demo:.4f}, крит.знач = {w_crit_demo:.4f}")
    print(f"  Разница статистик: {abs(t_stat_demo - w_stat_demo):.4f}")
    print()
```
