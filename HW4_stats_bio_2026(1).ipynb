Предположим, что вы разработали лекарство от коронавируса. Перед применением оно обязательно должно пройти клинические испытания. Для начала было разрешено проверить лекарство на двух независимых группах по 10 человек. Одна группа принимает плацебо, другая -- ваш препарат. Большее количество пациентов на первом этапе брать не разрешают -- слишком велики риски отрицательного результата.

Для каждого пациента измерялось количество дней от приема препарата до выздоровления. Получились следуюшие результаты:

```python
x = [6, 16, 8, 13, 9, 4, 7, 10, 3, 14]  # плацебо
y = [5, 10, 3, 1, 5, 3, 19, 2, 2, 5]  # лекарство
```

Что вы можете сказать на основе этих результатов?
* Лекарство эффективнее, подтверждается статистическими методами;
* Наверное, лекарство эффективнее, но статистическими методами это пока не подтверждено, нужно продолжить эксперимент. Подумайте, как обосновать необходимость продолжение эксперимента;
* По результатам эксперимента нельзя сделать какой-либо вывод. Стоит ли продолжать эксперименты? Если да, то четко это обоснуйте;
* Лекарство неэффективно, нужно немедленно прекращать эксперимент.

```python
import math

group_a = [6, 16, 8, 13, 9, 4, 7, 10, 3, 14]
group_b = [5, 10, 3, 1, 5, 3, 19, 2, 2, 5]

def get_stats(data):
    n = len(data)
    mean = sum(data) / n
    # Исправленная дисперсия (Bessel's correction: n-1)
    variance = sum((x - mean) ** 2 for x in data) / (n - 1)
    return mean, variance, n

# Получаем метрики
mean_a, var_a, n_a = get_stats(group_a)
mean_b, var_b, n_b = get_stats(group_b)

# Расчет t-статистики (формула Уэлча для независимых выборок)
t_stat = (mean_a - mean_b) / math.sqrt((var_a / n_a) + (var_b / n_b))

print(f"Группа A: Среднее = {mean_a:.2f}, Дисперсия = {var_a:.2f}")
print(f"Группа B: Среднее = {mean_b:.2f}, Дисперсия = {var_b:.2f}")
print(f"Разница средних: {mean_a - mean_b:.2f}")
print(f"Рассчитанное значение t-статистики: {t_stat:.3f}")
```

Значение t-критерия для уровня значимости a=0,05 меньше критического значения, поэтому нельзя сказать что лекарство эффективно. По дисперсии видно, что очень много шума, поэтому однозначный вывод по эим данным сделать нельзя. Нужно продолжать эксперимент и увеличить выборку для однозначного ответа.

#### 1. Зависимость распределения `DYRK1A_N` от генотипа

Проверьте гипотезу о равенстве средних с использованием критерия Стьюдента.  В реальных исследованиях нужно использовать конкретную альтернативу.

Сделайте выводы.

```python
# Разделяем выборки по генотипам для белка DYRK1A_N
group_control = data[data['Genotype'] == 'Control']['DYRK1A_N'].dropna()
group_trisomy = data[data['Genotype'] == 'Ts65Dn']['DYRK1A_N'].dropna()

t_stat, p_val = stats.ttest_ind(group_control, group_trisomy, equal_var=False)

print(f"Среднее (Control): {group_control.mean():.4f}")
print(f"Среднее (Ts65Dn): {group_trisomy.mean():.4f}")
print(f"t-статистика: {t_stat:.4f}")
print(f"p-value: {p_val:.4e}")
```

**Вывод**
Мы видим, что у мышей с синдромом Дауна (Ts65Dn) средний уровень экспрессии белка DYRK1A_N выше, чем у здоровых мышей, разница составляет около 11%.
p-value = 0.00245 значение меньше стандартного порога значимости α=0.05, следовательно вероятность того, что такая разница в средних получена случайно, составляет всего 0.24%. Это дает нам право уверенно отвергнуть нулевую гипотезу о равенстве средних.
t-статистика = -3.0372 отрицательное значение говорит о том, что среднее первой группы меньше среднего второй группы (Ts65Dn). Величина (−3.03) указывает на то, что разница между средними в 3 раза превышает стандартную ошибку этой разницы. Это считается сильным отклонением.

Посчитайте реализацию доверительного интервала величины сдвига.


```python
# Группы очищенные от пропусков
group_c = data[data['Genotype'] == 'Control']['DYRK1A_N'].dropna()
group_t = data[data['Genotype'] == 'Ts65Dn']['DYRK1A_N'].dropna()

# Разность средних
diff = group_c.mean() - group_t.mean()

# Построение доверительного интервала (95%)
# Используем объект DescrStatsW для удобного расчета интервала
import statsmodels.stats.api as sms

cm = sms.CompareMeans(sms.DescrStatsW(group_c), sms.DescrStatsW(group_t))
conf_int = cm.tconfint_diff(alpha=0.05, alternative='two-sided', usevar='unequal')

print(f"Точечная оценка сдвига: {diff:.4f}")
print(f"95% Доверительный интервал для сдвига: [{conf_int[0]:.4f}, {conf_int[1]:.4f}]")
```

#### 2. Зависимость распределения  `BDNF_N` от генотипа


Проверьте гипотезу о равенстве средних с использованием критерия Стьюдента при различных альтернативах, посчитайте оценку сдвига и доверительный интервал. Сделайте выводы.

```python
bdnf_control = data[data['Genotype'] == 'Control']['BDNF_N'].dropna()
bdnf_trisomy = data[data['Genotype'] == 'Ts65Dn']['BDNF_N'].dropna()

mean_c = bdnf_control.mean()
mean_t = bdnf_trisomy.mean()
shift = mean_c - mean_t

# Двусторонний тест
t_stat, p_two_sided = stats.ttest_ind(bdnf_control, bdnf_trisomy, equal_var=False)

# Односторонний тест (less — среднее первой группы МЕНЬШЕ, greater — БОЛЬШЕ)
# Мы проверяем гипотезу, что у здоровых (Control) BDNF выше (greater)
t_stat, p_greater = stats.ttest_ind(bdnf_control, bdnf_trisomy, equal_var=False, alternative='greater')

# Доверительный интервал для сдвига
cm = sms.CompareMeans(sms.DescrStatsW(bdnf_control), sms.DescrStatsW(bdnf_trisomy))
conf_int = cm.tconfint_diff(usevar='unequal')

print(f"Среднее (Control): {mean_c:.4f}")
print(f"Среднее (Ts65Dn): {mean_t:.4f}")
print(f"Оценка сдвига: {shift:.4f}")
print(f"P-value (двусторонний): {p_two_sided:.4e}")
print(f"P-value (Control > Ts65Dn): {p_greater:.4e}")
print(f"95% Доверительный интервал: [{conf_int[0]:.4f}, {conf_int[1]:.4f}]")
```

**Вывод:**

Средние значения: разница между 0.3211 и 0.3168 крайне мала.
Оценка сдвига: составляет всего 0.0043 это очень слабый эффект.

Если доверительный интервал разности средних содержит ноль, значит, мы не можем исключить ситуацию, при которой истинная разница между группами равна нулю.
Мы не можем отвергнуть нулевую гипотезус($H_0$). Различие в уровнях экспрессии белка BDNF_N между здоровыми и трисомными мышами на данной выборке статистически не значимо.
Возможно, уровень BDNF меняется не во всей коре, а только в определенных зонах (например, в гиппокампе), либо эффект проявляется только в определенном возрасте или после специфического обучения. На текущих данных строить диагностику по BDNF_N нельзя.

Как связаны доверительные интервалы с проверкой гипотез? Подтверждается ли это в данном примере?

**Ответ:**
Связь доверительных интервалов и проверки гипотез заключается в их математической эквивалентности: если заданный уровень значимости α (например, 0.05) не позволяет отвергнуть нулевую гипотезу, то соответствующий доверительный интервал (1−α) обязательно будет содержать внутри себя ноль.
В данном случае: для белка DYRK1A_N интервал не включал ноль и p-value был значимым, а для BDNF_N интервал пересек нулевую отметку, что согласуется с высоким p-value. Таким образом, оба метода — это разные способы описания одной и той же статистической уверенности в наличии или отсутствии эффекта.

**1.** **Оцените реальный уровень значимости** критерия для размеров выборки $1 \leqslant n \leqslant 50$ и визуализируйте полученную зависимость.

```python
# Параметры (уменьшили до 1000 для скорости)
alpha_nominal = 0.05
n_values = range(2, 51) # Начинаем с 2, так как для n=1 тест Стьюдента не работает
n_simulations = 1000

alpha_student = []
alpha_welch = []

for n in n_values:
    rejects_s = 0
    rejects_w = 0
    # Генерируем сразу все данные для этой итерации n (матрица n_simulations x n)
    # Это значительно ускоряет процесс
    data_x = np.random.normal(0, 1, (n_simulations, n))
    data_y = np.random.normal(0, 2, (n_simulations, n)) # Разные дисперсии!

    for i in range(n_simulations):
        # Тест Стьюдента
        _, p_s = stats.ttest_ind(data_x[i], data_y[i], equal_var=True)
        if p_s < alpha_nominal: rejects_s += 1

        # Тест Уэлча
        _, p_w = stats.ttest_ind(data_x[i], data_y[i], equal_var=False)
        if p_w < alpha_nominal: rejects_w += 1

    alpha_student.append(rejects_s / n_simulations)
    alpha_welch.append(rejects_w / n_simulations)

# Визуализация
plt.figure(figsize=(10, 5))
plt.plot(n_values, alpha_student, 'r-o', markersize=3, label='Стьюдент (ожидает равные дисперсии)')
plt.plot(n_values, alpha_welch, 'b-o', markersize=3, label='Уэлч (устойчив к разным дисперсиям)')
plt.axhline(y=0.05, color='green', linestyle='--', label='Целевой уровень 0.05')
plt.title('Реальный уровень значимости при нарушении условия равенства дисперсий')
plt.xlabel('Размер выборки (n)')
plt.ylabel('Вероятность ошибки I рода')
plt.legend()
plt.grid(True)
plt.show()
```

**Вывод:**
Критерий Уэлча немного лучше чем критерий Стюдента

**2.** Для значений $n \in \{5, 15, 30, 50, 100\}$ **оцените функцию мощности критерия** для значений $\theta \in (0, 5)$ по сетке с шагом 0.25 по методу Монте-Карло. Визуализируйте полученные зависимости. Посчитайте также мощность критерия по асимптотической формуле, полученной на лекции, и сравните с оценкой мощности методом Монте-Карло. Учтите, что асимптотическая формула содержит $n$, поэтому для нее тоже должно быть пять зависимостей.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Параметры
n_list = [5, 15, 30, 50, 100]
theta_grid = np.arange(0, 5.25, 0.25)
alpha = 0.05
n_sims = 2000 # Количество симуляций Монте-Карло

plt.figure(figsize=(14, 8))

for n in n_list:
    mc_powers = []
    asy_powers = []

    for theta in theta_grid:
        # --- МЕТОД МОНТЕ-КАРЛО ---
        # Генерируем выборки с разницей средних равной theta
        # H0: mu_x = mu_y vs H1: mu_x < mu_y (или mu_y - mu_x = theta)
        x = np.random.normal(0, 1, (n_sims, n))
        y = np.random.normal(theta, 1, (n_sims, n))

        # Считаем t-test для каждой симуляции (используем Уэлча для общности)
        _, p_vals = stats.ttest_ind(x, y, axis=1, equal_var=False, alternative='less')
        power = np.mean(p_vals < alpha)
        mc_powers.append(power)

        # --- АСИМПТОТИЧЕСКАЯ ФОРМУЛА ---
        # Стандартная ошибка разности средних при sigma=1: sqrt(1/n + 1/n) = sqrt(2/n)
        se = np.sqrt(2 / n)
        z_crit = stats.norm.ppf(1 - alpha)
        # Формула: P(Z > z_crit - theta/se)
        asy_p = 1 - stats.norm.cdf(z_crit - theta / se)
        asy_powers.append(asy_p)

    # Визуализация
    line, = plt.plot(theta_grid, mc_powers, 'o', label=f'MC (n={n})', markersize=4)
    plt.plot(theta_grid, asy_powers, '--', color=line.get_color(), alpha=0.7, label=f'Asy (n={n})')

plt.title('Мощность критерия: Монте-Карло vs Асимптотика')
plt.xlabel('Величина эффекта (theta)')
plt.ylabel('Мощность (1 - beta)')
plt.axhline(0.8, color='grey', linestyle=':', label='Желаемая мощность 0.8')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**Вывод:**
Результаты наглядно показывают проблемку малых выборок — если эффект θ мал, а данных мало, вероятность совершить ошибку II рода крайне высока.

**3.** Пусть $n=50$. **Сравните мощности критериев**, соответствующих двусторонней, левосторонней и правосторонней альтернативам при $\theta \in (0, 5)$ по сетке с шагом 0.25.

```python
# Параметры
n = 50
theta_grid = np.arange(0, 5.25, 0.25)
alpha = 0.05
n_sims = 5000

powers_two_sided = []
powers_greater = []
powers_less = []

for theta in theta_grid:
    # Генерируем выборки: X ~ N(0, 1), Y ~ N(theta, 1)
    # Тестируем H0: mu_x = mu_y
    x = np.random.normal(0, 1, (n_sims, n))
    y = np.random.normal(theta, 1, (n_sims, n))

    # Вычисляем p-value для разных альтернатив
    # Greater: H1: mu_x < mu_y (что соответствует нашему положительному theta)
    # Less: H1: mu_x > mu_y
    _, p_two = stats.ttest_ind(x, y, axis=1, alternative='two-sided')
    _, p_greater = stats.ttest_ind(x, y, axis=1, alternative='less')
    _, p_less = stats.ttest_ind(x, y, axis=1, alternative='greater')

    powers_two_sided.append(np.mean(p_two < alpha))
    powers_greater.append(np.mean(p_greater < alpha))
    powers_less.append(np.mean(p_less < alpha))

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(theta_grid, powers_greater, 'b-o', label='Правосторонняя (greater)', markersize=4)
plt.plot(theta_grid, powers_two_sided, 'g-s', label='Двусторонняя (two-sided)', markersize=4)
plt.plot(theta_grid, powers_less, 'r-^', label='Левосторонняя (less)', markersize=4)

plt.axhline(alpha, color='black', linestyle='--', alpha=0.5, label=f'Alpha = {alpha}')
plt.title(f'Сравнение мощности критериев при n={n}')
plt.xlabel('Величина сдвига (theta)')
plt.ylabel('Мощность')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

**Вывод:**
Правосторонняя альтернатива (синяя линия) ожидаемо показывает самую высокую мощность. Поскольку мы тестируем положительный сдвиг, вся критическая область сосредоточена в правом хвосте, что позволяет критерию быстрее и чаще фиксировать реальные различия. При θ≈1 его мощность уже достигает максимума (1.0). Двусторонняя альтернатива (зеленая линия) идет ниже синей. Чтобы сохранять готовность обнаружить сдвиг в любую сторону, этот критерий вынужден распределять уровень значимости α по обоим хвостам. Из-за этого он требует более сильного отклонения данных для того, чтобы признать результат значимым, и его мощность на малых θ всегда будет ниже, чем у правильно выбранного одностороннего теста.Левосторонняя альтернатива (красная линия) имеет практически нулевую мощность. Так как истинный сдвиг происходит вправо, а критерий «ищет» его слева, вероятность отклонения H0​ падает ниже уровня α. Это наглядно демонстрирует риск: если вы ошиблись с направлением гипотезы в одностороннем тесте, вы не заметите даже огромный реальный эффект.

Критерий Уэлча дал лучшие результаты в контроле ошибки 1 рода. В то время как классический критерий Стьюдента при разных дисперсиях в выборках начинает ошибаться чаще, чем задано уровнем α=0.05 особенно на малых выборках, критерий Уэлча стабильно удерживает реальный уровень значимости вблизи номинального.

Анализ мощности показал, что оба критерия ведут себя предсказуемо при увеличении размера выборки (n) и величины эффекта (θ). Однако использование односторонней альтернативы (при условии верного выбора направления) значительно повышает мощность по сравнению с двусторонней.
