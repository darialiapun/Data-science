Тип вина указан в первом столбце. Для работы с табличными данными используйте библиотеку `pandas`.

Постройте доверительные интервалы для параметров сдвига каждого из столбцов, предполагая, что столбцы имеют нормальное распределение.
Требуется построить:
 * асимтотические доверительные интервалы при помощи центральной предельной теоремы;
 * точные неасимптотические при помощи распределений хи-квадрат, Стьюдента.


 Запишите их в виде таблицы.

```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.datasets import load_wine

# 1. Загружаем данные
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
# Добавляем столбец с типом вина (он обычно в data.target)
df['wine_type'] = data.target

# В датасете типы вин: 0, 1, 2. "Первый тип" в sklearn — это 0.
# Если в твоем файле типы 1, 2, 3 — просто замени 0 на 1 ниже.
wine_1 = df[df['wine_type'] == 0]

# Выбираем нужные столбцы
columns_to_analyze = [
    'alcalinity_of_ash',
    'nonflavanoid_phenols',
    'proanthocyanins',
    'hue'
]

results = []
alpha = 0.95

for col in columns_to_analyze:
    series = wine_1[col]
    n = len(series)
    mean = series.mean()
    std_fixed = series.std(ddof=1) # исправленное стандартное отклонение (S)

    # --- 1. Асимптотический интервал (ЦПТ) ---
    # По конспекту: X_ср +- z_{1+a/2} * S / sqrt(n)
    z_crit = stats.norm.ppf((1 + alpha) / 2)
    margin_asymp = z_crit * (std_fixed / np.sqrt(n))
    ci_asymp = f"({mean - margin_asymp:.3f}, {mean + margin_asymp:.3f})"

    # --- 2. Точный неасимптотический интервал (Стьюдент) ---
    # По конспекту: X_ср +- t_{1+a/2, n-1} * S / sqrt(n)
    t_crit = stats.t.ppf((1 + alpha) / 2, df=n-1)
    margin_exact = t_crit * (std_fixed / np.sqrt(n))
    ci_exact = f"({mean - margin_exact:.3f}, {mean + margin_exact:.3f})"

    results.append({
        "Параметр": col,
        "Асимптотический (ЦПТ)": ci_asymp,
        "Точный (Стьюдент)": ci_exact
    })

# Оформляем в виде таблицы
final_table = pd.DataFrame(results)
print("Доверительные интервалы для вина 1-го типа:")
print(final_table.to_string(index=False))
```

**2.** Пусть есть выборка $X_1, ..., X_n$. Опишите процедуру бутстрепа построения доверительного интервала для величины $\theta = \mathsf{E} X_1$. Рассмотрите три типа  бутстрепных доверительных интервалов.
Выпишите соответствующие формулы.

**Описание**:

Пусть имеется выборка $X_1, \dots, X_n$ из некоторого распределения с неизвестным математическим ожиданием $\theta = \mathbb{E}X_1$. Построим доверительный интервал для $\theta$ с уровнем доверия $\alpha$.

#### Общая процедура
1. Вычисляем точечную оценку параметра по исходной выборке: $\hat{\theta} = \bar{X} = \frac{1}{n}\sum_{i=1}^n X_i$.
2. Генерируем $B$ бутстрепных выборок $\mathbb{X}^*_b$ путем извлечения $n$ элементов из исходной выборки **с возвращением**.
3. Для каждой выборки вычисляем бутстрепную оценку $\hat{\theta}^*_b = \bar{X}^*_b$.
4. Получаем вариационный ряд бутстрепных оценок: $\hat{\theta}^*_{(1)} \le \hat{\theta}^*_{(2)} \le \dots \le \hat{\theta}^*_{(B)}$.


#### Три типа бутстрепных доверительных интервалов

**1. Метод перцентилей **
Основан на квантилях эмпирического распределения бутстрепных оценок.
$$\theta \in \left[ \hat{\theta}^*_{\left( B \cdot \frac{1-\alpha}{2} \right)} ; \quad \hat{\theta}^*_{\left( B \cdot \frac{1+\alpha}{2} \right)} \right]$$
*Применяется, когда распределение $\hat{\theta}$ симметрично относительно $\theta$.*

**2. Центральный базовый интервал **
Использует распределение разности $(\hat{\theta}^* - \hat{\theta})$ для аппроксимации распределения $(\hat{\theta} - \theta)$.
$$\theta \in \left[ 2\hat{\theta} - \hat{\theta}^*_{\left( B \cdot \frac{1+\alpha}{2} \right)} ; \quad 2\hat{\theta} - \hat{\theta}^*_{\left( B \cdot \frac{1-\alpha}{2} \right)} \right]$$
*Позволяет корректировать небольшое смещение в данных.*

**3. Bootstrap t-interval**
Использует бутстреп для моделирования распределения $t$-статистики. Требует оценки стандартной ошибки $se^*$ для каждой бутстрепной выборки.
Строится статистика: $t^*_b = \frac{\hat{\theta}^*_b - \hat{\theta}}{se^*_b}$.
$$\theta \in \left[ \hat{\theta} - t^*_{\left( B \cdot \frac{1+\alpha}{2} \right)} \cdot se ; \quad \hat{\theta} - t^*_{\left( B \cdot \frac{1-\alpha}{2} \right)} \cdot se \right]$$
*Является наиболее точным (второго порядка точности), но вычислительно более затратным.*

**3.** Реализуйте функцию вычисления выборки оценок $\theta^*_1, \ldots \theta^*_B$ для оценки $\widehat \theta$ методом бутстрепа. Функция должна принимать на вход выборку и количество бутстрепных выборок $B$.

Для ускорения расчетов можете использовать broadcasting (не является обязательным требованием). Учите, что в таком случае внутри функции у вас могут получиться очень большие матрицы, из-за которых может кончиться оперативная память.

```python
import numpy as np

def bootstrap(
    sample: np.ndarray,
    B: int = 100_000
) -> np.ndarray:

    n = len(sample)

    # Генерируем матрицу индексов размером (B, n)
    # Каждый индекс выбирается из диапазона [0, n-1] с возвращением
    indices = np.random.randint(0, n, size=(B, n))

    # Используем broadcasting для получения бутстрепных выборок
    # Мы превращаем индексы в реальные значения из sample
    bootstrap_samples = sample[indices]

    # Считаем оценку (в данном случае среднее) по каждой строке
    # axis=1 означает вычисление среднего вдоль каждой из B выборок
    bootstrap_estimates = np.mean(bootstrap_samples, axis=1)

    return bootstrap_estimates
```

**4.** Реализуйте три типа бутстрепных доверительных интервалы в виде функций, принимающих на вход выборку оценок, полученных с помощью бутстрепа. Реализовывать вычисление бутстрепных интервалов для подвыборок размера от 1 до n не нужно, только для всей выборки.

```python
import numpy as np
from typing import Tuple

def bootstrap_normal_confidence_interval(
    theta: float,
    bootstrap_estimations: np.ndarray,
    alpha: float = 0.95
) -> Tuple[float, float]:
    """
    Считает нормальный бутстрепный интервал.
    Использует стандартную ошибку бутстрепного распределения и квантиль N(0,1).
    """
    from scipy import stats

    # Стандартная ошибка — это стандартное отклонение бутстрепных оценок
    se = np.std(bootstrap_estimations, ddof=1)

    # Квантиль нормального распределения
    z_crit = stats.norm.ppf((1 + alpha) / 2)

    left = theta - z_crit * se
    right = theta + z_crit * se

    return left, right


def bootstrap_central_confidence_interval(
    theta: float,
    bootstrap_estimations: np.ndarray,
    alpha: float = 0.95
) -> Tuple[float, float]:
    """
    Считает центральный (базовый) бутстрепный интервал.
    Формула: [2*theta - theta*_(1-alpha/2), 2*theta - theta*_(alpha/2)]
    """
    # Вычисляем квантили бутстрепного распределения
    q_left = np.percentile(bootstrap_estimations, (1 - alpha) / 2 * 100)
    q_right = np.percentile(bootstrap_estimations, (1 + alpha) / 2 * 100)

    # Отражаем относительно theta
    left = 2 * theta - q_right
    right = 2 * theta - q_left

    return left, right


def bootstrap_quantile_confidence_interval(
    theta: float,
    bootstrap_estimations: np.ndarray,
    alpha: float = 0.95
) -> Tuple[float, float]:
    """
    Считает квантильный (перцентильный) бутстрепный интервал.
    Формула: [theta*_(alpha/2), theta*_(1-alpha/2)]
    """
    # Просто берем соответствующие перцентили из массива оценок
    left = np.percentile(bootstrap_estimations, (1 - alpha) / 2 * 100)
    right = np.percentile(bootstrap_estimations, (1 + alpha) / 2 * 100)

    return left, right
```

**5.** Для каждого признака постройте бутстрепные доверительные интервалы для $\theta = \mathsf{E} X_1$ и сравните их.

```python
# 1. Список белков, которые мы анализируем
proteins = ['ITSN1_N', 'DYRK1A_N', 'ELK_N']

# 2. Инициализируем таблицу только для этих столбцов
data = pd.DataFrame(
    index=['normal', 'central', 'quantile', 'theta', 'interval_length'],
    columns=proteins
)

alpha = 0.95
B = 10000

# 3. Запускаем расчет
for col in proteins:
    # Очищаем данные от NaN и берем значения
    sample = df[col].dropna().values

    # Точечная оценка
    theta_hat = np.mean(sample)

    # Генерация бутстрепных оценок (функция уже должна быть в памяти)
    boot_estimates = bootstrap(sample, B=B)

    # Расчет интервалов (функции уже должны быть в памяти)
    ci_norm = bootstrap_normal_confidence_interval(theta_hat, boot_estimates, alpha)
    ci_cent = bootstrap_central_confidence_interval(theta_hat, boot_estimates, alpha)
    ci_quan = bootstrap_quantile_confidence_interval(theta_hat, boot_estimates, alpha)

    # Заполнение
    data.loc['normal', col] = ci_norm
    data.loc['central', col] = ci_cent
    data.loc['quantile', col] = ci_quan
    data.loc['theta', col] = theta_hat
    data.loc['interval_length', col] = ci_quan[1] - ci_quan[0]

# 4. Вывод результата без лишних NaN
print("Итоговая таблица сравнения интервалов:")
display(data)
```

```python
def draw_confidence_interval(
    left: np.ndarray,
    right: np.ndarray,
    estimation: Optional[np.ndarray] = None,
    ylim: Tuple[Optional[float], Optional[float]] = (None, None),
    estim_label: str = '',
    color_interval: Optional[str] = None,
    interval_label: Optional[str] = None
):
    # Массив размеров выборки n от 1 до len(left)
    n_values = np.arange(1, len(left) + 1)

    plt.fill_between(n_values, left, right, alpha=0.2, color=color_interval, label=interval_label)

    if estimation is not None:
        plt.plot(n_values, estimation, color=color_interval, label=estim_label, linewidth=2)

    plt.ylim(ylim)
    plt.xlabel('Размер выборки (n)')
    plt.ylabel('Значение параметра')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)

# --- ВИЗУАЛИЗАЦИЯ ДЛЯ ПРИЗНАКОВ ---

N = 100
B = 1000 # Для графиков можно взять B поменьше, чтобы считалось быстрее
alpha = 0.95

for col in proteins:
    # Генерируем выборку X_1 ... X_N из исходных данных белка
    # (берем первые 100 доступных значений)
    full_sample = df[col].dropna().values[:N]

    # Списки для хранения границ в зависимости от n
    left_bounds = []
    right_bounds = []
    estimations = []

    # Для каждой подвыборки размера n от 1 до N
    for n in range(1, N + 1):
        subsample = full_sample[:n]
        theta_hat = np.mean(subsample)

        # Считаем бутстрепные оценки для текущей подвыборки
        boot_ests = bootstrap(subsample, B=B)

        # Используем, например, квантильный интервал
        l, r = bootstrap_quantile_confidence_interval(theta_hat, boot_ests, alpha)

        left_bounds.append(l)
        right_bounds.append(r)
        estimations.append(theta_hat)

    # Превращаем в массивы для отрисовки
    left_bounds = np.array(left_bounds)
    right_bounds = np.array(right_bounds)
    estimations = np.array(estimations)

    # Рисуем график
    plt.figure(figsize=(10, 6))
    draw_confidence_interval(
        left=left_bounds,
        right=right_bounds,
        estimation=estimations,
        estim_label=f'Оценка среднего ({col})',
        color_interval='blue',
        interval_label='95% бутстрепный интервал'
    )
    plt.title(f'Динамика доверительного интервала для {col}')
    plt.show()
```
